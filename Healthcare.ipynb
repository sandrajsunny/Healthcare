## Statistical Calculations

### Mean,median,variance,standard deviation,sum,maximum value,minimum value calculations


```python
#Mean,median,variance,standard deviation,sum,maximum value,minimun value calculations

#Maximum

max_patient_age = float(round(df['patient_age'].max(), 2))
max_patient_sat_score = float(round(df['patient_sat_score'].max(), 2))
max_patient_waittime= float(round(df['patient_waittime'].max(), 2))

#Minimum

min_patient_age = float(round(df['patient_age'].min(), 2))
min_patient_sat_score = float(round(df['patient_sat_score'].min(), 2))
min_patient_waittime= float(round(df['patient_waittime'].min(), 2))

#Sum

sum_patient_age = float(round(df['patient_age'].sum(), 2))
sum_patient_sat_score = float(round(df['patient_sat_score'].sum(), 2))
sum_patient_waittime= float(round(df['patient_waittime'].sum(), 2))

#Mean

avg_patient_age= float(round(df['patient_age'].mean(), 2))
avg_patient_sat_score = float(round(df['patient_sat_score'].mean(), 2))
avg_patient_waittime = float(round(df['patient_waittime'].mean(), 2))

#Median

median_patient_age = float(round(df['patient_age'].median(), 2))
median_patient_sat_score = float(round(df['patient_sat_score'].median(), 2))
median_patient_waittime	 = float(round(df['patient_waittime'].median(), 2))

#Variance

variance_patient_age = float(round(df['patient_age'].var(), 2))
variance_patient_sat_score = float(round(df['patient_sat_score'].var(), 2))
variance_patient_waittime = float(round(df['patient_waittime'].var(), 2))

#Standard Deviation

std_patient_age = float(round(df['patient_age'].std(), 2))
std_patient_sat_score = float(round(df['patient_sat_score'].std(), 2))
std_patient_waittime = float(round(df['patient_waittime'].std(), 2))


headers = ["Parameter", "patient_age", "patient_sat_score", "patient_waittime"]
 
data = [["Mean", avg_patient_age,avg_patient_sat_score, avg_patient_waittime],
        ["Median", median_patient_age,median_patient_sat_score,median_patient_waittime],
        ["Variance", variance_patient_age,variance_patient_sat_score,variance_patient_waittime],
        ["StandardDeviation", std_patient_age,std_patient_sat_score,std_patient_waittime],
        ["Min Value",min_patient_age,min_patient_sat_score ,min_patient_waittime],
        ["Max Value",max_patient_age,max_patient_sat_score,max_patient_waittime],
        ["Sum Value",sum_patient_age,sum_patient_sat_score,sum_patient_waittime]]

tabulate.PRESERVE_WHITESPACE = True
tabulate.WIDE_CHARS_MODE = False
print(tabulate(data, headers=headers, tablefmt="grid"))
```
